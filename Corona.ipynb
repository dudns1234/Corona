```python
new_col=covid.loc[covid['date_dt']=='2020-05-31',['location','total_deaths','total_cases','population','aged_65_older','aged_70_older','median_age','diabetes_prevalence','cvd_death_rate','female_smokers','male_smokers']]
```

```python
new_col
```

```python
def covid_cases_rate(total_cases, population):
    cases_rate = total_cases / population
    return cases_rate

covid_cases_rate(11468,51269183)  #우리나라 인구대비 확진자 비율
```

```python
cases_rate_list = []

for row in range(len(new_col)):
    record = new_col.iloc[row]
    total_cases = record['total_cases']
    population = record['population']
    cases_rate = covid_cases_rate(total_cases, population)
    cases_rate_list.append(cases_rate)
    
new_col['per_cases_rate'] = cases_rate_list

#new_col[['total_deaths', 'population', 'cases_rate']].head(20)
```

```python
def covid_deaths_rate(total_deaths, population):
    deaths_rate = total_deaths / population
    return deaths_rate

covid_deaths_rate(270,51269183)  #우리나라 인구대비 사망자 비율
```

```python
deaths_rate_list = []

for row in range(len(new_col)):
    record = new_col.iloc[row]
    total_deaths = record['total_deaths']
    population = record['population']
    deaths_rate = covid_deaths_rate(total_deaths, population)
    deaths_rate_list.append(deaths_rate)
    
new_col['per_deaths_rate'] = deaths_rate_list

#new_col[[ 'total_deaths', 'population', 'deathes_rate']].head(20)
```

```python
new_col['deaths_rate']=np.round(new_col['total_deaths']/new_col['total_cases'],2)
```
